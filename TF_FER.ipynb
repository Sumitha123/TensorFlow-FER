## Data pre-processing

```python
#Subtracting the mean value from each image
training_images = training_images - training_images.mean(axis=1).reshape(-1,1)

#Setting the image norm to be 100
training_images = np.multiply(images,100.0/255.0)
pixel_mean = training_images.mean(axis=0)
pixel_std = np.std(training_images, axis=0)
images = np.divide(np.subtract(training_images,pixel_mean), pixel_std)

print('\nImage shape : ',images.shape)

image_pixels = images.shape[1]
print('\nFlat pixel value : ',image_pixels)

image_width = image_height = np.ceil(np.sqrt(image_pixels)).astype(np.uint8)

print('\nImage width : ', image_width)

flat_labels = training_data["emotion"].values.ravel()
labels_count = np.unique(flat_labels).shape[0]
print('Number of facial expressions : ', labels_count)
#One hot encoding for emotion classes
def dense_to_one_hot_encoding(dense_labels, num_emotions):
    num_labels = dense_labels.shape[0]
    index_offset = np.arange(num_labels) * num_emotions
    labels_one_hot_encoding = np.zeros((num_labels, num_emotions))
    labels_one_hot_encoding.flat[index_offset + dense_labels.ravel()] = 1
    return labels_one_hot_encoding
labels = dense_to_one_hot_encoding(flat_labels, labels_count)
labels = labels.astype(np.uint8)
```

```python
# Splitting data into training & validation to reduce overfitting
validation_size = 1709
validation_images = images[:validation_size]
validation_labels = labels[:validation_size]
train_images = images[validation_size:]
train_labels = labels[validation_size:]
print('Number of images in final training dataset : ', (len(train_images)))
```

## Building a CNN Model

```python
import tensorflow as tf
# Removing previous Tensors and Operations
tf.reset_default_graph()

#Initialization of weights
def weight_var(shape):
    initial = tf.truncated_normal(shape, stddev=1e-4)
    return tf.Variable(initial)

def bias_var(shape):
    initial = tf.constant(0.1, shape=shape)
    return tf.Variable(initial)

# Constructing a two-dimensional convolutional layer. 
def conv2d(x, W, padd):
    return tf.nn.conv2d(x, W, strides=[1, 1, 1, 1], padding=padd)

# Constructs a two-dimensional pooling layer using the max-pooling algorithm.
def max_pool_2x2(x):
    return tf.nn.max_pool(x, ksize=[1, 3, 3, 1], strides=[1, 2, 2, 1], padding='SAME')

# Local layer weight initialization
def local_weight_var(shape):
    initial = tf.truncated_normal(shape, stddev=0.04)
    return tf.Variable(initial)

def local_bias_var(shape):
    initial = tf.constant(0.0, shape=shape)
    return tf.Variable(initial)
```

### TensorBoard Support
#### Scalars - Accuracy, Weights, Biases, Cross-entropy, Drop-out
#### Images - Convolutional layer outputs, Convolutional layer weight visualization
#### Graph - Full network graph
#### Distributions and Histograms - Weights and Biases, Activations

```python
print('\nBuilding the Deep Neural Network..')
with tf.name_scope("Input_layer") as scope:
    # Creating placeholder for image data
    x = tf.placeholder('float', shape=[None, image_pixels])
    # Creating placeholder for labels
    y_ = tf.placeholder('float', shape=[None, labels_count])
    

# First convolutional layer
with tf.name_scope("Convolutional_layer_1") as scope:
    with tf.name_scope('Weights'): 
        W_conv1 = weight_var([5, 5, 1, 64])
        tf.summary.histogram('Histogram', W_conv1)
    with tf.name_scope('Biases'):
        bias_conv1 = bias_var([64])
        tf.summary.histogram('Histogram', bias_conv1)
    with tf.name_scope('Input_reshape'):
        # Reshaping the image data 
        image = tf.reshape(x, [-1,image_width , image_height,1])
        tf.summary.image('image', image,10)
    with tf.name_scope('Activations'):
        h_conv1 = tf.nn.relu(conv2d(image, W_conv1, "SAME") + bias_conv1)
        tf.summary.histogram('Histogram', h_conv1)
    with tf.name_scope('Max_pool'):
        h_pool1 = max_pool_2x2(h_conv1)
        h_norm1 = tf.nn.lrn(h_pool1, 4, bias=1.0, alpha=0.001/9.0, beta=0.75)
    with tf.name_scope('Image_output_conv1'):
        image = h_conv1[0:1, :, :, 0:16]
        image = tf.transpose(image, perm=[3,1,2,0])
        tf.summary.image('Image_output_conv1', image)


# Second convolution layer
with tf.name_scope("Convolutional_layer_2") as scope:
    with tf.name_scope('Weights'):  
        W_conv2 = weight_var([5, 5, 64, 128])
        tf.summary.histogram('Histogram', W_conv2)
    with tf.name_scope('Biases'):
        bias_conv2 = bias_var([128])
        tf.summary.histogram('Histogram', bias_conv2)
    with tf.name_scope('Activations'):
        h_conv2 = tf.nn.relu(conv2d(h_norm1, W_conv2, "SAME") + bias_conv2)
        tf.summary.histogram('Histogram', h_conv2)
    with tf.name_scope('Max_pool'):
        h_norm2 = tf.nn.lrn(h_conv2, 4, bias=1.0, alpha=0.001/9.0, beta=0.75)
        h_pool2 = max_pool_2x2(h_norm2)
    with tf.name_scope('Image_output_conv2'):               
        image = h_conv2[0:1, :, :, 0:16]
        image = tf.transpose(image, perm=[3,1,2,0])
        tf.summary.image('Image_output_conv2', image)

            
# Fully connected layers
with tf.name_scope("Fully_connected_layer_3") as scope:
    with tf.name_scope('Weights'):
        W_fc1 = local_weight_var([12 * 12 * 128, 3072])
        tf.summary.histogram('Histogram', W_fc1)
    with tf.name_scope('Biases'):
        b_fc1 = local_bias_var([3072])
        tf.summary.histogram('Histogram', b_fc1)
    with tf.name_scope('Max_pool'):
        h_pool2_flat = tf.reshape(h_pool2, [-1, 12 * 12 * 128])
    with tf.name_scope('Activations'):
        h_fc1 = tf.nn.relu(tf.matmul(h_pool2_flat, W_fc1) + b_fc1)
        tf.summary.histogram('Histogram', h_fc1)
                   
            
with tf.name_scope("Fully_connected_layer_4") as scope:
    with tf.name_scope('Weights'):
        W_fc2 = local_weight_var([3072, 1536])
        tf.summary.histogram('Histogram', W_fc2)
    with tf.name_scope('Biases'):
        b_fc2 = local_bias_var([1536])
        tf.summary.histogram('Histogram', b_fc2)
    with tf.name_scope('Activations'):
        h_fc2_flat = tf.reshape(h_fc1, [-1, 3072])
        h_fc2 = tf.nn.relu(tf.matmul(h_fc2_flat, W_fc2) + b_fc2)
        tf.summary.histogram('Histogram', h_fc2)

    # Dropout
    keep_prob = tf.placeholder('float')
    h_fc2_drop = tf.nn.dropout(h_fc2, keep_prob)
# Readout layer for deep net
with tf.name_scope("Readout_layer_3") as scope:    
    with tf.name_scope('Weights'):
        W_fc3 = weight_var([1536, labels_count])
        tf.summary.histogram('Histogram', W_fc3)
    with tf.name_scope('Biases'):
        b_fc3 = bias_var([labels_count])
        tf.summary.histogram('Histogram', b_fc3)
            
#Softmax layer
with tf.name_scope("Softmax_layer") as scope:
    y = tf.nn.softmax((tf.matmul(h_fc2_drop, W_fc3) + b_fc3))
```

```python
# Settings for the training
LEARNING_RATE = 0.0001
with tf.name_scope("Optimizer") as scope:
    # Cost function
    cross_entropy = -tf.reduce_sum(y_*tf.log(y))
    tf.summary.scalar("Error", cross_entropy)
    # Optimization using Adam optimizer
    train_step = tf.train.AdamOptimizer(LEARNING_RATE).minimize(cross_entropy)
    
# Evaluation of the predicted labels
correct_prediction = tf.equal(tf.argmax(y,1), tf.argmax(y_,1))

accuracy = tf.reduce_mean(tf.cast(correct_prediction, 'float'))

# Prediction function
predict = tf.argmax(y,1)

# Set training iterations to 5000
TRAINING_ITERATIONS = 5000
DROPOUT = 0.5
BATCH_SIZE = 50

epochs_completed = 0
index_in_epoch = 0
num_examples = train_images.shape[0]


# Serving data by batches
def next_batch(batch_size):
    global train_images
    global train_labels
    global index_in_epoch
    global epochs_completed

    start = index_in_epoch
    index_in_epoch += batch_size

    # When all training data has been used, it is reordered randomly
    if index_in_epoch > num_examples:
        # Finished epoch
        epochs_completed += 1
        # Shuffle the data
        perm = np.arange(num_examples)
        np.random.shuffle(perm)
        train_images = train_images[perm]
        train_labels = train_labels[perm]
        # Start next epoch
        start = 0
        index_in_epoch = batch_size
        assert batch_size <= num_examples
    end = index_in_epoch
    return train_images[start:end], train_labels[start:end]
print('DNN successfully built..')
```

## Training the Model

```python
# Starting TensorFlow session
#init = tf.initialize_all_variables()
init = tf.global_variables_initializer()
sess = tf.InteractiveSession()
sess.run(init)
summaryMerged = tf.summary.merge_all()
filename = "./summary_log/run" + datetime.datetime.now().strftime("%Y-%m-%d--%H-%M-%s")
writer = tf.summary.FileWriter(filename, sess.graph)
# Visualisation variables
train_accuracies = []
validation_accuracies = []
x_range = []
display_step=1
```

```python
tf.reset_default_graph()
for i in range(TRAINING_ITERATIONS):

    # Fetching new batch
    batch_xs, batch_ys = next_batch(BATCH_SIZE)

    # Checking accuracy on every 1st,2nd,...,10th,20th,...,100th... step
    if i % display_step == 0 or (i + 1) == TRAINING_ITERATIONS:

        train_accuracy = accuracy.eval(feed_dict={x: batch_xs,
                                                  y_: batch_ys,
                                                  keep_prob: 1.0})
        if (validation_size):
            validation_accuracy = accuracy.eval(feed_dict={x: validation_images[0:BATCH_SIZE],
                                                           y_: validation_labels[0:BATCH_SIZE],
                                                           keep_prob: 1.0})
            print('Training accuracy :  %.2f  Validation accuracy :  %.2f  for step %d' % (
            train_accuracy, validation_accuracy, i))

            validation_accuracies.append(validation_accuracy)

        else:
            print('Training accuracy :  %.4f for step %d' % (train_accuracy, i))
        train_accuracies.append(train_accuracy)
        x_range.append(i)

        # Increasing display_step
        if i % (display_step * 10) == 0 and i and display_step < 100:
            display_step *= 10
    # Training on batch
    _,sumOut = sess.run([train_step,summaryMerged], feed_dict={x: batch_xs, y_: batch_ys, keep_prob: DROPOUT})
    writer.add_summary(sumOut, i)
```
